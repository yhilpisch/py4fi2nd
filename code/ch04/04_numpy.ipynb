### Speed Comparison

```python
import random
I = 5000
```

```python
%time mat = [[random.gauss(0, 1) for j in range(I)] \
             for i in range(I)]  
```

```python
mat[0][:5]  
```

```python
%time sum([sum(l) for l in mat])  
```

```python
import sys
sum([sys.getsizeof(l) for l in mat])  
```

## Memory Layout

Cf. https://eli.thegreenplace.net/2015/memory-layout-of-multi-dimensional-arrays/

```python
x = np.random.standard_normal((1000000, 5))  
```

```python
y = 2 * x + 3  
```

```python
C = np.array((x, y), order='C')  
```

```python
F = np.array((x, y), order='F')  
```

```python
x = 0.0; y = 0.0  
```

```python
C[:2].round(2)  
```

```python
%timeit C.sum()  
```

```python
%timeit F.sum()  
```

```python
%timeit C.sum(axis=0)  
```

```python
%timeit C.sum(axis=1)  
```

```python
%timeit F.sum(axis=0)  
```

```python
%timeit F.sum(axis=1)  
```

```python
F = 0.0; C = 0.0  
```
